```python
import numpy as np
```

```python
start_point = np.array([-50, -50, 0]).astype('float')
print(start_point)
```

```python
x_length = 200
y_length = 100
z_length = 50
interval = 0.5
```

```python
end_point_wp = start_point + np.array([x_length, y_length, start_point[2]])
end_point_volume = start_point + np.array([x_length, y_length, z_length])
print(end_point_wp, end_point_volume)
```

```python
num_waypoints = 4*y_length + 2
print(num_waypoints)
num_y_patterns = int((num_waypoints-2)/4)
print(num_y_patterns)
```

```python
x_pattern = np.tile(np.array([end_point_wp[0], end_point_wp[0], start_point[0], start_point[0]]), num_y_patterns)
print(x_pattern, x_pattern.shape)
```

```python
x = x_pattern.copy()
x = np.insert(x, 0, start_point[0])
x = np.append(x, end_point_wp[0])
print(x, x.shape)
```

```python
y = np.array([[i, i] for i in np.arange(start_point[1], end_point_wp[1]+interval, interval)]).flatten()
print(y, y.shape)
```

```python
z = np.array([start_point[2]]*num_waypoints)
print(z, z.shape)
```

```python
xyz_wp = np.array([x, y, z]).T
print(xyz_wp, xyz_wp.shape)
```

```python
z = np.arange(start_point[2], end_point_volume[2]+interval, interval)
print(z, z.shape)
```

```python
volume = np.zeros((z.shape[0], x.shape[0], 3))
print(volume.shape)
```

```python
for h in range(volume.shape[0]):
    new_z = xyz_wp.copy()
    new_z[:, 2] = z[h]
    volume[h] = new_z
print(volume.shape)
```
